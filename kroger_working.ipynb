```python
def get_product_token():
    base = 'https://api-ce.kroger.com/v1/connect/oauth2/token'
    clientinfo = "Z3JvY2VyeWJ1ZGR5LTllNDg2Y2UzYWE1MGE2MzMyZTE0ZjRiODVhYTBjNjQwNDI4MjkzNzEwNzU3NzY1NzU1NTp1YUpyVGRJM2pLU3JJbTNpX1U4X2NJYmZJTk5XRVRhRW9xVGFQdGRD"
    headers = {
        'Content-Type':'application/x-www-form-urlencoded',
        'Authorization': f'Basic {clientinfo}',
        'scope':'product.compact',
        'grant_type':'client_credentials'

    }
    payload = {
        'grant_type' : 'client_credentials',
        'scope' : 'product.compact'
    }
#     resp = requests.post(base,data = payload, headers=headers)
    resp = requests.post(base, data=payload,headers=headers)
    return resp.json()
```

```python
def get_locations(zipcode = "37209"):
    base = f"https://api-ce.kroger.com/v1/locations?filter.zipCode.near={zipcode}"
    token = get_location_token().get('access_token')
    headers = {
        'Authorization': f'bearer {token}',
        "Content-Type": "application/json; charset=utf-8"
    }
    response = requests.get(base, headers = headers)
   # return response
    stores_df = pd.DataFrame(response.json().get('data'))
    return stores_df
```

```python
def get_products(locationId = '02600547', searchterm = 'milk', brand = 'Kroger'):
    base = f"https://api-ce.kroger.com/v1/products?filter.brand={brand}&filter.term={searchterm}&filter.locationId={locationId}"
    token = get_product_token().get('access_token')
    headers = {
        "Accept" : "application/json",
        "Authorization": f"Bearer {token}"
    }
    response = requests.get(base, headers = headers)
    return pd.DataFrame(response.json().get('data'))
```
